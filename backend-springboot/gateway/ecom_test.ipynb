```typescript

const gatewayUrl = "http://localhost:5160"


async function fetchWithHandling(url, options) {
  try {
    const response = await fetch(url, options);

    if (!response.ok) {
      // Optionally, parse error body for more details
      let errorText = await response.text();
      let errorBody;
      try {
        errorBody = JSON.parse(errorText);
      } catch {
        errorBody = errorText;
      }

      // in case the response body is empty
      if (!errorBody) {
        errorBody = 'No error message returned from server';
        }

      throw new Error(`result = HTTP ${response.status}: ${errorBody}`);
    }

    const body = await response.json().catch(() => null);

    return body;

  } catch (err) {
    // Handle network errors or thrown errors above
    console.error('Fetch error:', err);
    throw err;
  }
}
```
