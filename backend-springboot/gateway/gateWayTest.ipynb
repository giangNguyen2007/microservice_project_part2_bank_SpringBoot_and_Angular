```typescript


const credentials ={
    name: "giangNguyen",
    email : "rairacer@gmail.comn",
    password: "myPassword"
}

async function login()  {

    const response = await fetch("http://localhost:8085/user/login", {
        method: "POST",
        headers: { "Content-Type": "application/json" },
        body: JSON.stringify(credentials),
    });

    if (response.status !== 200) {
        throw new Error("Login failed");
    }

    const body = await response.json().catch(() => null);

    if (!body?.token) {
        throw new Error("missing token in response");
    }

    if (!body?.userId) {
        throw new Error("missing userId in response");
    }

    console.log("Login OK, token:", body.token);

    return body;
                                                                                                                                                                                                                                                                                                                                                                                                        
}
```

```typescript
// define fetch function

// const endpointAccount = "http://localhost:8085/account";

// const endpointTransaction = "http://localhost:8085/transaction";

// const endpointLogin = "http://localhost:8085/user/login";


const bankGateway_url = "http://localhost:8085"


async function fetchWithHandling(url, options) {
  try {
    const response = await fetch(url, options);

    if (!response.ok) {
      // Optionally, parse error body for more details
      let errorText = await response.text();
      let errorBody;
      try {
        errorBody = JSON.parse(errorText);
      } catch {
        errorBody = errorText;
      }

      // in case the response body is empty
      if (!errorBody) {
        errorBody = 'No error message returned from server';
        }

      throw new Error(`result = HTTP ${response.status}: ${errorBody}`);
    }

    const body = await response.json().catch(() => null);

    return body;

  } catch (err) {
    // Handle network errors or thrown errors above
    console.error('Fetch error:', err);
    throw err;
  }
}
```
